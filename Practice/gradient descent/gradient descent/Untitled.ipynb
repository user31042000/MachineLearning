```python
import numpy as np
import math
```

```python
def gradient_descent(x,y):
    mcur=0
    bcur=0
    n = len(x)
    rate = 0.01
    iteration = 100
    costcal = 0
    for i in range(iteration):
        
        y_pred = mcur*x+bcur
        cost = (1/n) *sum((y-y_pred)**2)
        
        if math.isclose(costcal, cost):
            print(f"m {mcur} , b {bcur} ,  iteration {i} , cost {cost}  ")
            break
        costcal = cost
        
        md= -(2/n)*sum(x*(y-y_pred))
        bd= -(2/n)*sum((y-y_pred))
        mcur = mcur - rate*md
        bcur = bcur - rate*bd
        print(f"m {mcur} , b {bcur} ,  iteration {i} , cost {cost}  ")
```
