```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = np.array([1,2,3,4,5])
price= np.array([10,20,30,40,50])
```

```python
x_train = area
y_train = price
```

```python
m = x_train.shape[0]
print(m)
```

```python
plt.scatter(area,price,marker= "x",c ="r")
plt.xlabel("area")
plt.ylabel("price")
plt.title("graph")
```

```python
i = 0 # Change this to 1 to see (x^1, y^1)

x_i = area[i]
y_i = price[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
w= 10
b = 0
```

```python
def model(x,w,b):
    m = x.shape[0]
    fwb = np.zeros(m)
    for i in range(m):
        fwb[i] = w*x[i]+b
    return fwb
```

```python
twb = model(x_train,w,b)
plt.plot(area,twb,c = "b",label="predicted")
plt.scatter(area,price,marker= "x",c ="r",label ="actual")
plt.xlabel("area")
plt.ylabel("price")
plt.title("graph")
plt.legend()
plt.show
```
